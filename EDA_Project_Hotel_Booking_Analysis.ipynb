## 3. ***Data Wrangling***

### Data Wrangling Code

Dropping duplicates

````python
# Dropping the duplicate values
hotel_booking_df.drop_duplicates(inplace = True)
````

````python
# after removing duplicate , count of total rows
print(f'After removing duplicates , number of rows are :  {hotel_booking_df.shape[0]}')
````

Replacing null values

````python
# droping the company column as it  has highest null values
hotel_booking_df.drop(['company'], axis=1, inplace=True)

# replacing null values in children column with 0  as there are only 4 missing values ,assuming that family had 0 children
# replacing null values in agent column with 0 assuming those rooms were booked without company/agent
hotel_booking_df[['children','agent']] = hotel_booking_df[['children','agent']].fillna(0)
````

````python
# replacing null values in country column as 'Others' assuming that user didnt fill their country name
hotel_booking_df['country'].fillna('Others', inplace = True)
````

````python
# checking that now there are no null values
hotel_booking_df.isna().sum().sort_values(ascending=False)
````

````python
#Some rows have a total of zero adults, children, or babies, indicating that no booking was made.
#Therefore, such rows can be removed.
````

````python
hotel_booking_df.drop(hotel_booking_df[hotel_booking_df['adults']+hotel_booking_df['babies']+hotel_booking_df['children'] == 0].index, inplace = True)
````

````python
# number of rows having with 0 in adult babies and children coloumn
zero_entry=hotel_booking_df[hotel_booking_df['adults']+hotel_booking_df['babies']+hotel_booking_df['children'] == 0].shape[0]
print(f"There are {zero_entry} rows which has zero value")
````

## changing data type

````python
#showing the info of the data to check datatype
hotel_booking_df.info()
````

````python
# We have seen that childern & agent column as datatype as float whereas it contains only int value,so change datatype as int64
hotel_booking_df[['children', 'agent']] = hotel_booking_df[['children', 'agent']].astype('int64')
````

````python
# for our understanding, in column 'is_canceled' we will replace the value from (0,1) to not_canceled, is canceled.
hotel_booking_df['is_canceled'] = hotel_booking_df['is_canceled'].replace([0,1], ['not canceled', 'is canceled'])
````

````python
#Same for 'is_repeated_guest' column
hotel_booking_df['is_repeated_guest'] = hotel_booking_df['is_repeated_guest'].replace([0,1], ['not repeated', 'repeated'])
````

## addition of new coloumns

````python
#total stay in nights
# We created a column for total stays in nights by combining the weekday and weekend stay columns.
hotel_booking_df['total_stay_in_nights'] = hotel_booking_df ['stays_in_week_nights'] + hotel_booking_df ['stays_in_weekend_nights']
````

````python
# We have created a col for revenue using total stay * adr
hotel_booking_df['revenue'] = hotel_booking_df['total_stay_in_nights'] *hotel_booking_df['adr']
````

````python
# Also, for information, we will add a column with total number of guests coming for each booking
hotel_booking_df['total_guest'] = hotel_booking_df['adults'] + hotel_booking_df['children'] + hotel_booking_df['babies']
````

### What all manipulations have you done and insights you found?

**1**)Dropping duplicates: There were a total of 31,994 duplicate entries. To clean the data, we removed these duplicates, leaving 87,396 rows after the process.


**2**)Handling Null Values: The columns 'children,' 'country,' 'agent,' and 'company' contain null values. The 'company' column, having the highest number of null values, will be dropped.



*   Then replacing null values in agent column with 0 assuming those rooms were booked without company/agent

*   Then I have filled  null values in children column with 0  as there are only 4 missing values ,assuming that family had 0 children.
*   Replacing null values in the country column with 'others'to avoid confusion.




**3**) Some rows have a total of zero adults, children, or babies, indicating no booking was made. As a result, we removed 166 rows where the values in the adults, babies, and children columns were zero.

**4**) The data types of the 'children' and 'agent' columns were set to float, but they only contain integer values. Therefore, we changed their data types to int64.

**5**) For clarity, we will replace the values in the 'is_canceled' column from (0, 1) to 'not_canceled' and 'is_canceled.' The same will be applied to the 'is_repeated_guest' column.

**6)**There are few columns required in Data for analysis purpose which is created from the given columns.


Total Guests: This column will assist us in evaluating the total number of guests and revenue. We calculate this value by adding the total number of adults, children, and babies.

Revenue: We determine revenue by multiplying the average daily rate (adr) by the total number of guests. This column will be used to analyze the profit and growth of each hotel.

#### Chart - 10

**What is ADR accross different months?**

````python
#  Using groupby funtion
bookings_by_months_df = hotel_booking_df.groupby(['arrival_date_month', 'hotel'])['adr'].mean().reset_index()

# Create month list
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# It will take the order of the month list in the dataframe along with values
bookings_by_months_df['arrival_date_month'] = pd.Categorical(bookings_by_months_df['arrival_date_month'], categories = months, ordered = True)

# Sorting values
bookings_by_months_df = bookings_by_months_df.sort_values('arrival_date_month')
````

````python
# Setting the chart size
plt.figure(figsize=(15,5))

# Plotting the values in a line chart
sns.lineplot(x=bookings_by_months_df['arrival_date_month'],y=bookings_by_months_df['adr'],hue=bookings_by_months_df['hotel'])

# Setting the labels and title
plt.title('ADR across each month', fontsize=20)
plt.xlabel('Month Name', fontsize=12)
plt.ylabel('ADR', fontsize=12)

# Show chart
plt.show()
````

##### 1. Why did you pick the specific chart?

I used a line plot in this case because it effectively illustrates the trend of ADR over time, making it easy to observe how ADR fluctuates or changes across different time periods.

#### Chart - 15 - Correlation Heatmap

````python
numeric_columns = hotel_booking_df.select_dtypes(include=['int64', 'float64']).columns

# Calculating the correlation matrix
correlation_matrix = hotel_booking_df[numeric_columns].corr()

# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

````

##### 1. Why did you pick the specific chart?

I chose a heatmap to visualize the correlation matrix because it effectively represents the relationships among multiple numeric variables in a dataset. Heatmaps offer a clear and intuitive visual representation of how these variables correlate with one another, using color gradients to indicate the strength and directionwhether positive or negativeof these correlations.

##### 2. What is/are the insight(s) found from the chart?

Total Stay in Nights is a central variable that shows a strong correlation with both weeknight and weekend stays, significantly influencing revenue. Revenue is notably affected by factors such as ADR, total stay in nights, and the total number of guests. In contrast, variables like lead time and previous cancellations exhibit little to no correlation with most other variables, suggesting their limited impact on other factors.
